# Ordinal Encode - Ordinal Encode (Independent Feat)

```python
# Reading Data into a frame
import pandas as pd
df = pd.read_excel('data\p03.xlsx')
df = df.dropna()
df.head()

from sklearn.preprocessing import OrdinalEncoder

custom_order = ["Bronze", "Silver", "Gold"]


# Create an instance of OrdinalEncoder with the custom order
oe = OrdinalEncoder(categories=[custom_order])

# Fit and transform the encoding on the "Size" column
df["Medal"] = oe.fit_transform(df[["Medal"]])
```

```python
oe.inverse_transform([[0],[1],[2]])
```

# OHE - Nominal Data

```python
# Count of Total Unique Cities 
df.City.nunique()
```

Select Nominal Column -> Label Encode -> One Hot Encode

```python
# Defiine OHE and Encode a column
# Label Encode first-> Extract features -> Convert to df
ohe = preprocessing.OneHotEncoder(drop='first')
df['City']= le.fit_transform(df['City'])
encoded_features = ohe.fit_transform(df[['City']]).toarray()
```

```python

# Create a DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names(['City']))
encoded_df
```

```python

df_encoded = pd.concat([df, encoded_df], axis=1)
df_encoded.head()
```

# Multi Category

```python
# Viewing unique categories
df.Team.nunique()
```

```python
# Filtering to only 10 categories

top_10 = [x for x in df.Team.value_counts().sort_values(ascending=False).head(9).index]
top_10
```

```python
# Re-define categories

df.Team  = df.Team.apply(lambda x: x if x in top_10 else 'other' )
```

```python
# Viewing unique categories again
df.Team.nunique()
```

```python
# Label Encode first-> Extract features -> Convert to df

df['Team']= le.fit_transform(df['Team'])
encoded_features = ohe.fit_transform(df[['Team']]).toarray()
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names(['City']))
```

```python
df_encoded = pd.concat([df, encoded_df], axis=1)
df_encoded.head()
```
